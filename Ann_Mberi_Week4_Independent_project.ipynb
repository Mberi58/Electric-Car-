###Data Cleaning

```python
#Since our data frames have alot of columns and information we need to do some data cleaning. We go through the five stages when cleaning the data
#Validity
#Accuracy 
#Completeness
#Consistency
#Uniformity
```

####A. Validity

```python
#We drop the data we are not using in our analysis This involves dropping irrelevant data which we won't use in our analysis

#First we drop the unnecessary column in our analysis that should be the geo point column
df.drop("Geo point", axis=1, inplace=True)
```

```python
# The ID column and the public column are similar hence we drop the ID column
df.drop("ID", axis=1, inplace=True)
```

```python
# In our analysis the display will not be of much use, so we drop it
df.drop("Displayed comment", axis=1, inplace=True)
```

```python
# Finally we drop the last column which is the scheduled column, it still not necessary 
df.drop("Scheduled at", axis=1, inplace=True)
```

```python
#We know that the data was collected from ist April to 9th April so I create a date column. Then after we will
#concatenate the year, month and day columns into one
df.loc[:,'Date'] = pd.to_datetime(df.day.astype(str) + '-' + df.month.astype(str) + '-' + df.year.astype(str))
df
```

```python
# We then drop the day, month, year and minute column to make the data logical 
# Remember in our analysis question we were asked about the hour so we will not drop it
df.drop(["year","month","day","minute"], axis=1, inplace=True)
```

```python
#Run the data to view the remaining columns 
df
```

####B.Accuracy

```python
# This involves checking whether the resources in our data file are accurate
# i.e cars, bluecar counter, utilib, utilib 1.4, charging slots and slots
df["Cars"].sum()
```

```python
# The total number of cars is 11668
#The Bluecar counter = Cars. Therefore the sum is equal to
#Proof below by running the sum 
df["Bluecar counter"].sum() 
```

```python
# On checking the accuracy we run both the sum of the cars and that of the blue cars 
Accuracy = df["Cars"].sum() == df["Bluecar counter"].sum()
Accuracy
```

####C.Completeness

```python
#We confirm if the data file is complete and we do that by checking for missing or null values 
df.isnull().any()
```

####D.Consistency

```python
#We confirm if there are any duplicate values
df.duplicated().any() 
```

```python
#Please remember that the car column and the blue car column are the same so we drop the car column
df.drop("Cars", axis=1, inplace=True) 
```

####E. Uniformity

```python
# For uniformity we make sure that all the column names are the uniform
df = df.rename(columns = str.lower)
df.head()
```
